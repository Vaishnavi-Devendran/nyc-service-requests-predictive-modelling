```python
from scipy.stats import f_oneway
import pandas as pd

anova_results = {}
for col in categorical_cols:
    if sample_df[col].nunique() > 1:
        try:
            groups = [g['Resolution Time (minutes)'].dropna() for _, g in sample_df.groupby(col)]
            if len(groups) >= 2:
                f_val, p_val = f_oneway(*groups)
                anova_results[col] = p_val
        except Exception:
            pass

anova_df = pd.DataFrame.from_dict(anova_results, orient='index', columns=['p_value']).sort_values('p_value')
print("\nTop 10 categorical features by ANOVA significance:")
print(anova_df.head(10))
```

```python
import pandas as pd
import numpy as np
import unicodedata
import re

# 1. Normalize Resolution Description text
def normalize_resolution_text(s):
    if pd.isna(s):
        return ""
    s = unicodedata.normalize("NFKC", str(s))   # fix weird quotes etc
    s = s.lower().strip()
    return s

df_selected['Resolution_norm'] = df_selected['Resolution Description'].apply(normalize_resolution_text)

# 2. Define safe, intersecting patterns
patterns = {
    # duplicate complaints (HPD / DEP etc)
    'res_duplicate_building': r'duplicate of a building[- ]wide condition',
    'res_duplicate_other': r'duplicate of a previously filed complaint|original complaint is being addressed',

    # inspection workflow
    'res_inspection_in_progress': r'inspection is currently in progress|will inspect the condition',
    'res_inspection_attempted': r'attempted to conduct an inspection|attempted to conduct the inspection',
    'res_inspection_done': r'conducted an inspection|inspected this condition|inspected the condition',

    # access issues
    'res_no_access': r'not able to gain access|unable to gain access|unable to complete the inspection|unable to gain entry',

    # no condition / no violation found
    'res_no_condition': r'found no condition at the location|didn\'?t observe a violation|no violation at the location|no evidence of the violation',

    # referred or reassigned
    'res_referred': r'referred this complaint|referred it to|referred this complaint to|required re-assignment to a new dot unit',

    # work order created / filed
    'res_work_order': r'work order has been filed|created a work order|opened a repair order',

    # online status info (DOT)
    'res_status_online': r'service request status .* department of transportation.*website|click the "learn more" link|click the learn more link',
}

# 3. Create binary columns from patterns
for col, pat in patterns.items():
    df_selected[col] = df_selected['Resolution_norm'].str.contains(pat, regex=True, na=False).astype(int)
```

```python
import lightgbm as lgb

train_data = lgb.Dataset(train[features], label=train[target])
test_data  = lgb.Dataset(test[features],  label=test[target])

params = {
    'objective': 'regression',
    'metric': 'rmse',
    'learning_rate': 0.05,
    'num_leaves': 40,
    'feature_fraction': 0.8,
    'bagging_fraction': 0.8,
    'bagging_freq': 4,
    'min_data_in_leaf': 50
}

model = lgb.train(
    params,
    train_data,
    valid_sets=[train_data, test_data],
    valid_names=['train', 'test'],
    num_boost_round=2000,
    callbacks=[lgb.early_stopping(stopping_rounds=100)]
)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error

# Predict log1p values
y_pred_log = model.predict(test[features], num_iteration=model.best_iteration)

# Convert back to minutes
y_pred_minutes = np.expm1(y_pred_log)
y_true_minutes = np.expm1(test['log_Resolution_Time'])

# Real-world RMSE
rmse_minutes = mean_squared_error(y_true_minutes, y_pred_minutes)**0.5
print("RMSE (minutes):", rmse_minutes)
```
